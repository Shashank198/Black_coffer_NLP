```python
def extract_text_from_url(url):
    # Make a GET request to the URL
    response = requests.get(url)
    
    # Parse the HTML content using Beautiful Soup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Extract the text from the HTML
    text = soup.get_text()
    # Tokenize the text string
    tokens = nltk.word_tokenize(text)

# Remove stop words from the tokens list
    filtered_tokens = [token for token in tokens if not token.lower() in stop_words]

# Join the filtered tokens into a new string
    clean_text = ' '.join(filtered_tokens)
    
    return clean_text
print(extract_text_from_url('https://mrec.ac.in/'))
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from nltk.corpus import cmudict
nltk.download('cmudict')
d = cmudict.dict()
# Define the dataframe of URLs
# urls_df = pd.DataFrame({
#     'url': ['https://en.wikipedia.org/wiki/Python_(programming_language)',
#             'https://en.wikipedia.org/wiki/Beautiful_Soup_(HTML_parser)',
#             'https://en.wikipedia.org/wiki/Web_scraping']
# })

# Define a function to extract text from a URL using Beautiful Soup
def extract_text_from_url(url):
    # Make a GET request to the URL
    response = requests.get(url)
    
    # Parse the HTML content using Beautiful Soup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Extract the text from the HTML
    text = soup.get_text()
    tokens = nltk.word_tokenize(text)
    sen=nltk.sent_tokenize(text)
    
# Remove stop words from the tokens list
    filtered_tokens = [token for token in tokens if not token.lower() in stop_words]
    
# Join the filtered tokens into a new string
    clean_text = ' '.join(filtered_tokens)
    
    return [tokens,sen]    
#     return text

# Iterate through the dataframe and extract text from each URL
personal_pronouns = ['i', 'you', 'he', 'she', 'it', 'we', 'they', 'me', 'him', 'her', 'us', 'them', 'mine', 'yours', 'his', 'hers', 'its', 'ours', 'theirs', 'myself', 'yourself', 'himself', 'herself', 'itself', 'ourselves', 'themselves']

for index, row in data_set.iterrows():
    url = row['URL']
    token= extract_text_from_url(url)
    tokens=token[0]
    sen=token[1]
    with open('positive-words.txt','r') as f:
        ps=f.read().splitlines()
    with open('negative-words (1).txt','r') as f1:
        ns=f1.read().splitlines()
    Positive_Score=sum(1 for word in tokens if word.lower() in ps)
#     op['POSITIVE SCORE'][index]=positive_score
    Negative_Score=sum(1 for word in tokens if word.lower() in ns)
    Polarity_Score = (Positive_Score-Negative_Score)/ ((Positive_Score + Negative_Score) + 0.000001)
    Subjectivity_Score = (Positive_Score + Negative_Score)/ ((len(tokens)) + 0.000001)
    Average_Sentence_Length = len(tokens)/ len(sen)
    
    SYLLABLE_PER_WORD=[]
# Count the number of complex words
    num_complex_words = 0
    for word in tokens:
      if(word.lower() in d):
        # Get the number of syllables in the word using the CMU dictionary
        num_syllables = [len(list(y for y in x if y[-1].isdigit())) for x in d[word.lower()]][0]
        if num_syllables >= 3:
            num_complex_words += 1
        SYLLABLE_PER_WORD.append(num_syllables)
    Percentage_of_Complex_words = num_complex_words /len(tokens)
    Fog_Index = 0.4 * (Average_Sentence_Length + Percentage_of_Complex_words)
    Average_Number_of_Words_Per_Sentence = len(tokens)/len(sen)
    COMPLEX_WORD_COUNT=num_complex_words
    WORD_COUNT=len(tokens)
    SYLLABLE_PER_WORD=sum(SYLLABLE_PER_WORD)//len(SYLLABLE_PER_WORD)
#     PERSONAL_PRONOUNS=
    num_personal_pronouns = sum(1 for word in tokens if word.lower() in personal_pronouns)
    AVG_WORD_LENGTH=sum(len(word) for word in tokens)
    
    with open('Op.csv.csv','a',newline='') as file:
        writer = csv.writer(file)
#     l=
#         writer.writerow(['Positive_Score', 'Negative_Score', 'Polarity_Score',
#        'Subjectivity_Score', 'Average_Sentence_Length',
#        'Percentage_of_Complex_words', 'Fog_Index',
#        'Average_Number_of_Words_Per_Sentence', 'COMPLEX_WORD_COUNT', 'WORD_COUNT',
#        'SYLLABLE_PER_WORD', 'PERSONAL_PRONOUNS', 'AVG_WORD_LENGTH'])#['Sentences Count', 'Total Words', 'Avg Word Length', 'Complex Words Count', 'Personal Pronouns Count'])
        writer.writerow([Positive_Score, Negative_Score, Polarity_Score,
       Subjectivity_Score, Average_Sentence_Length,
       Percentage_of_Complex_words, Fog_Index,
       Average_Number_of_Words_Per_Sentence, COMPLEX_WORD_COUNT, WORD_COUNT,
       SYLLABLE_PER_WORD, num_personal_pronouns, AVG_WORD_LENGTH])
#     print(Positive_Score,Negative_Score,Polarity_Score,Subjectivity_Score,Average_Sentence_Length,Percentage_of_Complex_words,Fog_Index,Average_Number_of_Words_Per_Sentence,COMPLEX_WORD_COUNT,WORD_COUNT,SYLLABLE_PER_WORD,num_personal_pronouns,AVG_WORD_LENGTH) 
    # Print the text to the console
#     print(f"Text for {url}:")
#     print(positive_score)
```
